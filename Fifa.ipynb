```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Fifa.csv")
df.head()
```

```python
sns.jointplot(x="Age", y="Potential", joint_kws={"alpha":0.1,"s":3,"color":"blue"} , data=df , marginal_kws={"color":"green"})
plt.show()
```

```python
sns.lmplot(x='Age', y="SprintSpeed",data=df,scatter_kws={"alpha":0.1,"s":3,"color":"red"},line_kws={"color":"blue"},lowess=True)
```

```python
sns.lmplot(x="BallControl",y="Dribbling",data=df,
           col="Preferred Foot",scatter_kws={"alpha":0.1,"color":"green"},
           line_kws={"color":"red"})
plt.show
```

```python
plt.figure(figsize=(20,8))
sns.boxplot(x="Overall",y="Age",hue="Preferred Foot",
            data=df ,
            palette="rocket")
plt.show()
```
